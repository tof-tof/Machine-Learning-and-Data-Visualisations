## Line charts

In this section, we'll be looking at the variations in the number of visitors to a selection of museums

```python
#setup
museum_filepath = "museum_visitors.csv"
museum_data = pd.read_csv(museum_filepath, index_col="Date", parse_dates=True)
```

```python
def drawLineChart(datas,title,xlabel,ylabel,width,height,data_col=None):
    sns.set_style("whitegrid")
    plt.figure(figsize=(width,height))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    sns.lineplot(data=datas,label=data_col)
    
drawLineChart(museum_data,"Monthly Visitors to a selection of Museums",'Date','Number of visitors',14,7)   
    
```

The plot shows the Avila Adobe generally has the most visitors, but there is an incredible spike of visitors to the Firehouse Museum around late summer in 2014, which is probably due to a bug event than run that year. 

```python
drawLineChart(museum_data['Avila Adobe'],"Monthly Visitors Avila Adobe",'Date','Number of visitors',14,7,data_col='Avila Adobe')   
```

Zooming in on just Avila Adobe allows us to notice that there is a classical time series trend, with the number of visitors dipping around fall and winter months but increasing in spring and summer. 

## Barcharts and Heatmaps

In this section, we'll be assesing how Genre and Platform of various games affects their rating

```python
#setup
ign_filepath = "ign_scores.csv"
ign_data = pd.read_csv(ign_filepath, index_col='Platform')
ign_data
```

```python
def drawBarChart(xdata,ydata,title,xlabel,ylabel,width,height):
    sns.set_style("ticks")
    plt.figure(figsize=(width,height))
    sns.barplot(y=ydata, x=xdata)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    
drawBarChart( ign_data['Racing'],ign_data.index,'average scores for racing games by platform', 'Platform','Rating',16,8)
```

From this bar chart, we can see the older,less popular platforms like Wii,Gameboy and Nintendo, tend to have slightly lower ratings, while newer, popular platforms like Xbox favour a little better 

```python
def drawHeatmap(plot_data,xlabel,title,width,height):
    plt.figure(figsize=(width,height))
    sns.heatmap(data=plot_data, annot=True)
    plt.xlabel(xlabel)
    plt.title(title)

drawHeatmap(ign_data,'Genre','Average Game Score by Platform and Genre',16,10)
```

the heatmap shows that once both platform and genre are considered, the popularity varies massively amongst these two factors. Xbox One tends to be the most popular platform, and Action Adventure the most popluar Genre

# Scatter Plots

In this section we are finding trends in data collected on different types of candy

```python
#setup
candy_filepath = "candy.csv"
candy_data = pd.read_csv(candy_filepath, index_col='id')
candy_data.head()
```


First we pose the question: Do people tend to prefer candies with higher sugar content?  


```python
def drawScatterplot(xdata,ydata,hueData=None):
    sns.set_style("white")
    sns.scatterplot(x=xdata, y=ydata, hue=hueData)
drawScatterplot(candy_data['sugarpercent'],candy_data['winpercent'])
```

The plot shows there is no strong correlation between the two variables, so sugar does not by an obvious role in popularity

```python
def drawRegplot(xdata,ydata,hue=None):
    sns.regplot(x=xdata, y=ydata)
drawRegplot(candy_data['sugarpercent'],candy_data['winpercent'])
```

Although there is no obvious correlation regression plot shows that overall, there could be a *slight* preferance of the more sugary candies, so one should probably take a closer look

```python
drawScatterplot(candy_data['sugarpercent'],candy_data['winpercent'],candy_data['chocolate'])
```

With the chocolate hue, we can see patterns beginning to emerge with chocolate candies, tending to be the pricer, but more popluar candies. This calls for further investigation

```python
def drawlmplot(x_col,y_col,hue_col,data):
    sns.lmplot(x=x_col,y=y_col,hue=hue_col, data=data) 
    
drawlmplot('pricepercent','winpercent','chocolate',candy_data)
```

From this plot we can see for chocolate candies, poplarity increases with price, but for non-chocolate candies, popularity decreases with price

```python
def drawCatScatplot(xdata, ydata):
    sns.swarmplot(x=xdata,y=ydata)
drawCatScatplot(candy_data['chocolate'],candy_data['winpercent'])
```

From the catergorical scatter plot, we can clearly see that chocolate candies tend to be more popular than candies without chocolate.

## Distributions

In this section, we'll be looking at data collected from microscopic images of breast cancer tumors, which are labelled benign or malignant


```python
#setup
cancer_b_filepath = "cancer_b.csv"
cancer_m_filepath = "cancer_m.csv"
cancer_b_data = pd.read_csv(cancer_b_filepath, index_col='Id')
cancer_m_data =  pd.read_csv(cancer_m_filepath, index_col='Id')
```

```python
sns.set_style("darkgrid")
sns.distplot(a=cancer_b_data['Area (mean)'],label='benign',kde=False) 
sns.distplot(a=cancer_m_data['Area (mean)'],label='malignant',kde=False)
plt.title('Average Tumor Areas')
plt.ylabel('frequency')
plt.legend()
```

from the histogram above, we can see that, Magliant tumors have higher areas on average and much more variation in size

```python
sns.set_style("dark")
sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label="Benign")
sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label="Malignant") # Your code here (benign tumors)
plt.xlabel('Radius (worst)')
plt.ylabel('frequency (percentage) of tumors in range')
plt.legend()
plt.title('Radii of benign and malignant tumors')
```

The KDE plot tells a similar story to the histogram, so if the tumor has a radius over 25, it is most like the tumor is maligant
